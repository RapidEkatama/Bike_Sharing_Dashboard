```python
#Copying the original dataset
df_day_copy = df_day.copy()
df_hour_copy = df_hour.copy()
```

```python
#Make a dictionary to change the categorical value from numeric into string 
Season_Map = {1:'Spring',
              2:'Summer',
              3:'Fall',
              4:'Winter'}

#Make a variable contain some columns
Day_Season = df_day_copy[['season','weathersit','temp','atemp','hum','windspeed']]
#Mapping the 'season' from numeric into string
Day_Season['season'] = Day_Season['season'].map(Season_Map)
Day_Season = Day_Season.groupby(['season']).mean()

Day_Season
```

From table above, we can see that each season has the own of characteristic. From `Fall` Season, it has the higher temperature (in Celcius degree). In `Spring` Season slightly has the lower humidity which is affect on more dry humidity that the other season.

```python
# Get the mean each customer in fall season
Fall_Day = df_day_copy[df_day_copy['season'] == 3]
Fall_Day_C = Fall_Day['casual'].mean().round(1)
Fall_Day_R = Fall_Day['registered'].mean().round(1)
Fall_Day_T = Fall_Day['cnt'].mean().round(1)

# Get the mean each customer in spring season
Spring_Day = df_day_copy[df_day_copy['season'] == 1]
Spring_Day_C = Spring_Day['casual'].mean().round(1)
Spring_Day_R = Spring_Day['registered'].mean().round(1)
Spring_Day_T = Spring_Day['cnt'].mean().round(1)

# Show the table contain of mean each group of customer
A = pd.DataFrame({'Season':['Fall','Spring','Mean Total'],
                  'Casual':[Fall_Day_C,Spring_Day_C,df_day_copy['casual'].mean().round(1)],
                  'Registered':[Fall_Day_R,Spring_Day_R,df_day_copy['registered'].mean().round(1)],
                  'Total Cust':[Fall_Day_T,Spring_Day_T,df_day_copy['cnt'].mean().round(1)],
                  })
A = A.style.set_caption('Table Mean of Customer Based on Differences in Temperature and Humidity Conditions')
A
```

Continuing the previous statement, from the different `Fall` and `Spring` season conditions compared to other seasons, it was concluded that in `Fall` Season has the higher casual and registered users and `Spring` has the lowest.

```python
#make a dictionary to change 'weekday' value from numeric into string
Day_Map = {0:'Sunday',
           1:'Monday',
           2:'Tuesday',
           3:'Wednesday',
           4:'Thursday',
           5:'Friday',
           6:'Saturday'}

#Mapping the dictionary into 'weekday' columns
df_day_copy['weekday'] = df_day_copy['weekday'].map(Day_Map)
df_day_copy['season'] = df_day_copy['season'].map(Season_Map)
df_day_copy
```

```python
#make a line plot from 'weekday' and 'cnt' (total users)
plt.figure(figsize=(8,5))
sns.lineplot(data=df_day_copy, x='weekday', y='cnt', errorbar=None)
plt.title('Chart of Total Users in One Year each Weekday')
```

From chart above, we can get a conclusion that in `Sunday` has the lowest total user, and in `Friday` it has the highest total user. With that information we can make a plan that keep an eye on customer in weekends and make a trial and error plan or evaluation plan in `Sunday` because in that day has the lowest Total Users and first weekday too.  

```python
#make a line plot from 'weekday' and 'cnt' (total users), and separate with 'season' category
plt.figure(figsize=(8,5))
sns.lineplot(data=df_day_copy, x='weekday', y='cnt', hue='season', errorbar=None)
plt.title('Chart of Total Users in One Year each Weekday Group by Season')
```

From chart above, we can conclude that in `Spring` has the lowest total users in any days. `Fall` season has the highest total users, and a slightly different in `Summer` and `Winter` season. On that chart we can make a evaluation because based on situation, how can be a bike sharing users in `Winter` season higher than users in `Spring` season.
